```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
images_path='output_videos/cropped_image.jpg'
image=cv2.imread(images_path)
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show
```

```python
#Taking Top Half of image
```

```python
top_half_image=image[0:int(image.shape[0]/2),:]
plt.imshow(top_half_image)
plt.show
```

# cluster the image into two cluster

```python
image_2d=top_half_image.reshape(-1,3)

kMeans=KMeans(n_clusters=2,random_state=0).fit(image_2d)

labels=kMeans.labels_

clusted_image=labels.reshape(top_half_image.shape[0],top_half_image.shape[1])

plt.imshow(clusted_image)
plt.show()
```

```python
corner_cluster=[clusted_image[0,0],clusted_image[0,-1],clusted_image[-1,0],clusted_image[-1,-1]]
non_player_cluster=max(set(corner_cluster),key=corner_cluster.count)
print(non_player_cluster)
```

```python
player_cluster=1-non_player_cluster
print(player_cluster)
```

```python
kMeans.cluster_centers_[player_cluster]
```
